# The substitution cipher

There are many contexts in which one would not want that a piece of information is available to others. In the modern world examples are the password of an email, the code of a locker or even a phone number. In fact, the need to protect information has long existed in humanity, and have decided the fate of history multiple times.

One of the first ways developed of securing information from unwanted eyes was to use a substitution cipher. In that method, the sender of a message tries to hide it's meaning by replacing each letter in the message by another predetermined letters. That process is called encryption and in that context, the original message is often referred to as the plain text, while the final message is referred as ciphertext.

In order to recover the message, the receiver of the encrypted message only needs to know which letter in the ciphertext corresponds to each letter in the plaintext (also know as the cipher key).

You can read more about substitution cipher here: https://en.wikipedia.org/wiki/Substitution_cipher

Your first task is to write a function substitution_cipher() that receive three string parameters: plaintext, alphabet, and key.

The function should perform a substitution encryption in the plaintext by replacing each letter in alphabet by the equivalent letter in key.

Ex.:
key = "kzhjcfdltbvwxmnairoyepgsuq"
alphabet = "abcdefghijklmnopqrstuvwxyz"

cipher_text = substitution_cipher('my encrypted message', alphabet, key)

cipher_text should be:
>XU CMHRUAOCJ XCYYKDC

(In this and other examples decoded texts are represented in lower case while encrypted text is in UPPER case)

(Note that each letter should have a unique corresponding key, otherwise the message might not be able to be retrieved later.)



````python
alphabet="abcdefghijklmnopqrstuvwxyz"
key = "kzhjcfdltbvwxmnairoyepgsuq".upper()

key= key.upper()
print(key)

substitution_cipher('my encrypted message', alphabet, key)
````

````python
def substitution_cipher(plain_text, alphabet, key):

  coded = ""
  for l in plain_text:
    if l in alphabet:
      coded+=key[alphabet.index(l)]
    else:
      coded+=l
  return coded
````

````python
def substitution_decode(cipher, alphabet, key):
  text = ''
  for l in cipher:
    if l in key:
      text += alphabet[key.index(l)]
    else:
      text += l

  return text
````

It may seem that this encryption method is secure if you don't know the key to decrypt. However, there are simple ways to find out the key based on statistics of the message. For instance, we can use frequency analysis to uncover what is the most probable correspondence between letters. This method was first introduced by the important arab matematician [Al-Kindi](
https://en.wikipedia.org/wiki/Al-Kindi), the father of the cryptography field and it consists of estimating the how often each letter appear in the message, and comparing it to how often each letter appear in the language in which the message is written. Of course this assumes that one knows the original language of the message!  

Now that you know how to implement a substitution cipher, let's try to create some functions to help us break the encryption.

First write a function that receives a str message and returns how many times each letter appear in the message. (You can return a list with the counts in alphabetical order, or a dictionary)



````python
def count_letters(text):
  alphabet = "abcdefghijklmnopqrtsuvwxyz"
  counts = []
  for l in alphabet:
    count = 0
    for c in text:
      if c==l:
        count+=1
    counts.append(count)
  return counts
````

Another important feature that one can look at in frequency analysis are which groups of two letters (bigram) occur more often. For example, in English the most common bigrams are 'th' and 'he', both present in the word 'the'. If someone sees the following encrypted message:

> JAD SUJ WY OLJ OF JAD GUE


A good guess would be that the 'ja' and 'ad' correspond to 'th' and 'eh'. Making possible to partially decrypt the message to:

>the SUt WY OLt OF the GUE

Of course, in very small texts like this one it's impossible to go much further using frequency analisis and decipher the original message ("the cat is out of the bag"). We would need more text to be able to link each letter in the cipher to its corresponding letter.

Write a function count_bigrams that receive a text and returns two lists: one with all the bigrams present in the text, and another list with their correspondent count.

Ex:



```python

bigrams = ['he', 'er', 're', 'an', 'nd', 'th']

```



````python
def count_bigrams(text):
  bigrams = []
  counts = []
  for word in text.split():
    for i in range(len(word)-1):
      bigram = word[i:i+2]
      if bigram in bigrams:
        i = bigrams.index(bigram)
        counts[i]+=1
      else:
        bigrams.append(bigram)
        counts.append(1)

  return bigrams, counts

bigrams, counts = count_bigrams("here and there")
print(bigrams)
print(counts)
# print(most_common(bigrams, counts, 3))
````

Finally, implement a helper function that receives a list of bigrams, their counts and an integer n. The function then returns the n most common bigrams

````python
def most_common(bigrams, counts, n):
  # Using list comprehension and lambda functions
  # r = [x for x,y in sorted(zip(bigrams, counts), key = lambda z: z[1], reverse = True)]
  # r = [:n]

  bigrams_ = bigrams.copy()
  # Using only for loops:
  r = []
  while len(r)<n:
    imax = 0
    max_count = 0
    for i in range(len(bigrams_)):
      if max_count<counts[i]:
        imax = i
        max_count = counts[i]
    r.append(bigrams_[imax])
    bigrams_.pop(imax)

  return r
````
