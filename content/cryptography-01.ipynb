# The substitution cipher

There are many contexts in which one would not want that a piece of information is available to others. In the modern world examples are the password of an email, the code of a locker or even a phone number. In fact, the need to protect information has long existed in humanity, and have decided the fate of history multiple times.

One of the first ways developed of securing information from unwanted eyes was to use a substitution cipher. In that method, the sender of a message tries to hide it's meaning by replacing each letter in the message by another predetermined letters. That process is called encryption and in that context, the original message is often referred to as the plain text, while the final message is referred as ciphertext.

In order to recover the message, the receiver of the encrypted message only needs to know which letter in the ciphertext corresponds to each letter in the plaintext (also know as the cipher key).

(You can read more about substitution cipher [here](https://en.wikipedia.org/wiki/Substitution_cipher)!)

# Frequency analysis

It may seem that this encryption method is secure if you don't know the key to decrypt. However, there are simple ways to find out the key based on statistics of the message. For instance, we can use frequency analysis to uncover what is the most probable correspondence between letters.

This method was first introduced by the important arab matematician [Al-Kindi](
https://en.wikipedia.org/wiki/Al-Kindi), the father of the cryptography field and it consists of estimating the how often each letter appear in the message, and comparing it to how often each letter appear in the language in which the message is written.
