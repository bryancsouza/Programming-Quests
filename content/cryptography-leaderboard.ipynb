from io import BytesIO

import requests
import pandas as pd
url = 'https://docs.google.com/spreadsheets/d/1oPKs0hOBGy4rRCftPseLYnYHCZgsyMYyaIa3kx2Ar78/edit?usp=sharing'
url = 'https://docs.google.com/spreadsheet/ccc?key=' + url.split('/')[-2] + '&output=csv'
r = requests.get(url)

data = r.content
df = pd.read_csv(BytesIO(data), index_col=0, parse_dates=[0], dayfirst=True)
# df['date'] = pd.to_datetime(df.index, dayfirst=True)
df['month'] = df.index.month
df['year'] = df.index.year
df.reset_index(drop=True, inplace=True)
df.iloc[:,0:2]

df.rename({'Name (alias)': 'Username'}, axis=1, inplace=True)
df["Date"] = df["month"].astype(str) + "/" + df["year"].astype(str) 


if df.loc[df["year"]==2025,["Username", "Program", "Date"]].size==0:
    print("No entries found..")
else:
    display(df.loc[df["year"]==2025,["Username", "Program", "Date"]])
