# Wall of Cryptoanalysts
Here you can find the names of the students who already completed this quest!


```python
from io import BytesIO

import requests
import pandas as pd

url = 'https://docs.google.com/spreadsheets/d/1EhGVBjtgeg1ZmknMi7b2orNKUSoShfzc0kxmRkFWFnI/edit?usp=sharing'
url = 'https://docs.google.com/spreadsheet/ccc?key=' + url.split('/')[-2] + '&output=csv'
r = requests.get(url)

data = r.content
df = pd.read_csv(BytesIO(data), index_col=0, parse_dates=[0])
# df['date'] = pd.to_datetime(df.index, dayfirst=True)
df['month'] = df.index.day
df['year'] = df.index.year
df.reset_index(drop=True, inplace=True)
df.iloc[:,0:2]

df.rename({'Name (alias)': 'Username'}, axis=1, inplace=True)
df["Date"] = df["month"].astype(str) + "/" + df["year"].astype(str) 
df.loc[:,["Username", "Program", "Date"]].tail(10)
```
